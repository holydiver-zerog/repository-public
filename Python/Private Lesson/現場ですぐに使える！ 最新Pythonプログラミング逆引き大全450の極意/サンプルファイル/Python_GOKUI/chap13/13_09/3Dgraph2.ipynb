```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d

def f(x, y):
    """xy2

    Args:
        x (ndarray): 2
        y (ndarray): 2

    Returns:
        ndarray: x**2 + y**2
    """
    return x**2 + y**2

# x1
x1 = np.arange(-3, 3, 0.25)
# x2
x2 = np.arange(-3, 3, 0.25)
# 2XY2
X, Y = np.meshgrid(x1, x2)
# X, Yfunc()
# Z
Z = f(X, Y)

# Figure
fig = plt.figure()
# Axes3D
ax = fig.add_subplot(111, projection='3d')
# x1
ax.set_xlabel("x1")
# x2
ax.set_ylabel("x2")
# x1x2 f(x1,x2)
ax.plot_wireframe(X,Y,Z)

plt.show()
```
