```python
# 3Dグラフの描画

import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d

# Figureを生成
fig = plt.figure()
ax = fig.add_subplot(
    111,            # 1行×1列の1番目のグリッドにAxesを追加
    projection='3d' # 3次元の投影面
    )

# テスト用のデータを取得
X, Y, Z = axes3d.get_test_data()

# 3Dワイヤーフレームをプロット
ax.plot_wireframe(X, Y, Z)

plt.show()
```
