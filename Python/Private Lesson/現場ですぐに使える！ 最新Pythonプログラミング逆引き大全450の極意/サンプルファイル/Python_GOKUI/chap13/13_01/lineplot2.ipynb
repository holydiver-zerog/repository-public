```python
import matplotlib.pyplot as plt

plt.plot([1, 2, 3, 4],      # xの値
        [1, 4, 9, 16],      # yの値
        linestyle='dotted', # ラインを点線にする
        linewidth=5,        # ライン幅は5pt
        color='red'         # ラインの色は赤
        )
plt.ylabel('y-label')       # y軸のラベルをプロット
plt.xlabel('x-label')       # x軸のラベルをプロット
plt.show()                  # グラフを表示
```
