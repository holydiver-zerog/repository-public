```python
import matplotlib.pyplot as plt
import numpy as np

t = np.arange(0.0, 2.0, 0.01)       # 02.00.01
s1 = np.sin(2*np.pi*t)              # (2t)
s2 = np.sin(4*np.pi*t)              # (4t)

# 21
plt.subplot(2, 1, 1, facecolor='pink')
plt.plot(t, s1)
# 21
plt.subplot(2, 1 ,2, facecolor='white')
plt.plot(t, s2)
plt.show()
```

```python
# 4
import matplotlib.pyplot as plt
import numpy as np

t = np.arange(0.0, 2.0, 0.01)            # 02.00.01
s1 = np.sin(2*np.pi*t)                   # (2t)
s2 = np.sin(4*np.pi*t)                   # (4t)
s3 = np.sin(6*np.pi*t)                   # (6t)
s4 = np.sin(8*np.pi*t)                   # (8t)

plt.subplot(2, 2, 1, facecolor='pink')   # 2211
plt.plot(t, s1)
plt.subplot(2, 2, 2, facecolor='white')  # 2212
plt.plot(t, s2)
plt.subplot(2, 2, 3, facecolor='silver') # 2221
plt.plot(t, s3)
plt.subplot(2, 2, 4, facecolor='pink')   # 2222
plt.plot(t, s4)
plt.show()
```

```python
# Figure
import matplotlib.pyplot as plt
import numpy as np

names = ['groupA', 'groupB', 'groupC'] # 
values = [1, 10, 100]                  # 

plt.figure(1,                          # Figureid
            figsize=(9, 3))            # 93

plt.subplot(1, 3, 1)                   # 
plt.bar(names, values)
plt.subplot(1, 3, 2)                   # 
plt.plot(names, values)
plt.subplot(1, 3, 3)                   # 
plt.scatter(names, values)
plt.suptitle('Three kinds of graphs')  # Figure

plt.show()
```

```python
# 
import matplotlib.pyplot as plt
import numpy as np

def damped (t):                    # 
    s1 = np.cos(2*np.pi*t)
    e1 = np.exp(-t)
    return s1 * e1

def undamped (t):                  # 
    return np.cos(2*np.pi*t)

t = np.arange(0.0, 5.0, 0.1)       # 0.05.00.1

plt.figure(1,                      # Figureid
            figsize=(10, 6))       # 93
            
plt.suptitle('Figure title', fontsize=16) # Figure

plt.subplot(1, 2, 1)               # 
plt.plot(t, damped(t))             # damped()
plt.xlabel('time (s)', size=12)    # x
plt.ylabel('Oscillation', size=12) # y
plt.title('Damped Oscillation')    # 

plt.subplot(1, 2, 2)               # 
plt.plot(t, undamped(t), '--')     # undamped()
plt.xlabel('time (s)', size=12)    # x
plt.ylabel('Oscillation', size=12) # y
plt.title('Undamped')              # 

plt.subplots_adjust(wspace=0.4,    # 
                    top=0.85)      # Figure
                    
plt.show()
```
