```python
# 単語の出現回数をカウントして頻度表を作る関数

def get_frequency(file):
    """単語の出現回数をカウントして頻度表を作る

    Args:
        file (str): ファイル名(パス)
    Returns:
        dict: 単語と出現回数
    """
    # 辞書を用意
    freq = {}
    # ファイルをオープン
    file_data = open(file, encoding='UTF-8')
    # ①段落を取り出す
    for line in file_data:
        words = line.split() # 単語に分割してリストにする(改行は除かれる)
        # ②リストから単語を取り出す
        for word in words:
            word = word.rstrip('.,:!?)’”') # 末尾のピリオド等を取り除く
            word = word.lstrip('(‘“')      # 冒頭のクォートを取り除く
            # ③辞書に単語と同じキーがあるか
            if word in freq:                
                freq[word] += 1 # キーの値に1加算
            # 該当するキーがなければ単語をキーにして値を1にする
            else:
                freq[word] = 1# 
    # 頻度表としての辞書を返す
    return freq
```

```python
# テキストファイルから単語の頻度表を作る

# ファイルを指定して頻度表の辞書を取得
freq = get_frequency('alice.txt')

# 取得した辞書から「英単語(頻度)」の形式で出力する
for word in freq: # 辞書のキー(単語)を取り出す
    # キー(単語)と値(頻度)を出力
    print(word + '(' + str(freq[word]) + ')' )
```

```python
# 単語の出現回数をカウントして頻度表を作る関数

def get_frequency(file):
    """単語の出現回数をカウントして頻度表を作る

    Args:
        file (str): ファイル名(パス)
    Returns:
        dict: 単語と出現回数
    """
    # 辞書を用意
    freq = {}
    # ファイルをオープン
    file_data = open(file, encoding='UTF-8')
    # ①段落を取り出す
    for line in file_data:
        words = line.split() # 単語に分割してリストにする(改行は除かれる)
        # ②リストから単語を取り出す
        for word in words:
            word = word.rstrip('.,:!?)’”') # 末尾のピリオド等を取り除く
            word = word.lstrip('(‘“')      # 冒頭のクォートを取り除く
            # ③辞書に単語と同じキーがあるか
            if word in freq:                
                freq[word] += 1 # キーの値に1加算
            # 該当するキーがなければ単語をキーにして値を1にする
            else:
                freq[word] = 1# 
    # 頻度表としての辞書を返す
    return freq
```

```python
# テキストファイルから単語の出現回数順に並べた頻度表を作る

# ファイルを指定して頻度表の辞書を取得
freq = get_frequency('alice.txt')

for word in sorted(
    freq,         # 対象の辞書
    key=freq.get, # 並べ替えの基準(key)を辞書の値にする
    reverse=True  # 降順で並べ替え
    ):
    # キー(単語)と値(頻度)を出力
    print(word + '(' + str(freq[word]) + ')')
```
